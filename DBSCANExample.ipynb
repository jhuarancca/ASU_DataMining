```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

data=np.loadtxt("casas.csv",delimiter=",")

clusters=DBSCAN(eps=2,min_samples=10).fit_predict(data)

#Plot
plt.figure(figsize=(10,10))
plt.scatter(data[:,0],data[:,1], c=clusters, s=100)
plt.xlabel("Antiguedad (Aos)")
plt.ylabel("Precio USD")
plt.box(False)
plt.show()

#https://www.youtube.com/watch?v=HMis89lGdkA
```
