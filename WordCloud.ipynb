```python
# install wordcloud
#!conda install -c conda-forge wordcloud==1.4.1 --yes
!conda install -c anaconda xlrd --yes
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')

# download file and save as alice_novel.txt
#!wget --quiet https://ibm.box.com/shared/static/m54sjtrshpt5su20dzesl5en9xa5vfz1.txt -O alice_novel.txt
#!wget --quiet https://github.com/jhuarancca/ASU_DataMining/edit/main/HuaranccaJuan_Tesis_maestria_2021.txt -O HuaranccaJuan_Tesis_maestria_2021.txt

#HuaranccaJuan_Tesis_maestria_2021.txt

# open the file and read it into a variable alice_novel
# alice_novel = open('alice_novel.txt', 'r').read()
fileWordCloud = open('HuaranccaJuan_Tesis_maestria_2021.txt', 'r').read()

print ('File downloaded and saved!')
```

```python
stopwords = set(STOPWORDS)
stopwords.add("y")
stopwords.add("de")
stopwords.add("en")
stopwords.add("la")
stopwords.add("que")
stopwords.add("para")
stopwords.add("el")
stopwords.add("desde")
stopwords.add("se")
stopwords.add("por")
stopwords.add("lo")
stopwords.add("las")
stopwords.add("del")
stopwords.add("su")
stopwords.add("los")
stopwords.add("con")
stopwords.add("año")
stopwords.add("un")
stopwords.add("como")
stopwords.add("al")
stopwords.add("fuente")
stopwords.add("elaboración")
stopwords.add("propia")
stopwords.add("son")
stopwords.add("una")
stopwords.add("e")
stopwords.add("es")
stopwords.add("sus")
stopwords.add("producto")
# instantiate a word cloud object
texto_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords

)

# generate the word cloud
texto_wc.generate(fileWordCloud)

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches # needed for waffle Charts

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the word cloud
plt.imshow(texto_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
from PIL import Image # converting images into arrays

# download image
#!wget --quiet https://ibm.box.com/shared/static/3mpxgaf6muer6af7t1nvqkw9cqj85ibm.png -O alice_mask.png
    
# save mask to alice_mask
texto_mask = np.array(Image.open('mapaMundi.png'))
    
print('Image downloaded and saved!')

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(texto_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()

# instantiate a word cloud object
alice_wc = WordCloud(background_color='white', max_words=2000, mask=texto_mask, stopwords=stopwords)

# generate the word cloud
alice_wc.generate(fileWordCloud)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
